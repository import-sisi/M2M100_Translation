```python
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer
```

```python
model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")
tokenizer = M2M100Tokenizer.from_pretrained("facebook/m2m100_418M")
```

```python
tokenizer.src_lang = "fa"
persian_text = "رئیس جمهور کشور به افغانستان سفر کرد"
encoded_persian = tokenizer(persian_text, return_tensors="pt")
generated_tokens = model.generate(**encoded_persian, forced_bos_token_id=tokenizer.get_lang_id("fr"))
```

```python
generated_tokens
```

```python
tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python

# Step 1: Persian (Farsi) to French
tokenizer.src_lang = "fa"
persian_text = "رئیس جمهور کشور به افغانستان سفر کرد"
encoded_persian = tokenizer(persian_text, return_tensors="pt")
generated_tokens_fr = model.generate(**encoded_persian, forced_bos_token_id=tokenizer.get_lang_id("fr"))

# Step 2: French to Japanese
tokenizer.src_lang = "fr"
generated_tokens_ja = model.generate(generated_tokens_fr, forced_bos_token_id=tokenizer.get_lang_id("ja"))

# Step 3: Japanese back to Persian
tokenizer.src_lang = "ja"
generated_tokens_fa = model.generate(generated_tokens_ja, forced_bos_token_id=tokenizer.get_lang_id("fa"))

# Decode all the translations at once using batch_decode
translations = tokenizer.batch_decode([generated_tokens_fr[0], generated_tokens_ja[0], generated_tokens_fa[0]], skip_special_tokens=True)

# Display the translations
print("French Translation:", translations[0])
print("Japanese Translation:", translations[1])
print("Back to Persian Translation:", translations[2])
```
